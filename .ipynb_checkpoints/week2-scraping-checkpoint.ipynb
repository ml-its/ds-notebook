````python
url = 'https://forlap.ristekdikti.go.id/mahasiswa/detailsemester/OTk0RkEzNTMtMEE1Ri00QjEzLThDOUMtNEQ3NUMyRDk3RTlG/20162'
req = requests.get('https://forlap.ristekdikti.go.id/mahasiswa/detailsemester/OTk0RkEzNTMtMEE1Ri00QjEzLThDOUMtNEQ3NUMyRDk3RTlG/20162', verify=False)
# throws ssl error,because pemerintah is males to update the damn TLS certificate
soup = BeautifulSoup(req.text,'lxml')

# print(soup.table)
print(soup.find('table'))

````
