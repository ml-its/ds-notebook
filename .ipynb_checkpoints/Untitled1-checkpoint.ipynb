```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.its.ac.id/matematika/dosen-staf/daftar-dosen/'
req = requests.get(url)
print(req.text)
```

```python
soup = BeautifulSoup(req.text, 'lxml')
soup.link
```

```python
semua_gambar = soup.find_all('img',{'class':'vc_single_image-img attachment-medium'})
```

```python
for dosen in semua_gambar: 
    print(dosen, '\n')
```

```python
for dosen in semua_gambar:
    print(dosen['src'])
```

```python
url = 'https://www.olx.co.id/all-results/q-mobil/'
req_paperwhite = requests.get(url)
soup_paperwhite = BeautifulSoup(req_paperwhite.text, 'lxml')
```

```python
seluruh_judul= soup_paperwhite.find_all('a', {'class':'marginright5 link linkWithHash detailsLink'})
```

```python
for paperwhite in seluruh_judul: 
    print(paperwhite['href'])
```

```python
seluruh_harga = soup_paperwhite.find_all('strong', class_='c000')
```

```python
for price in seluruh_harga: 
    print(price.text.strip())
```

```python
seluruh_kota = soup_paperwhite.find_all('p', class_='color-9 lheight14 margintop3')
seluruh_kota
```

```python
for city in seluruh_kota: 
    print(city.find('span').text)
```

```python
seluruh_tanggal = soup_paperwhite.find_all('td',class_='rel tcenter')
```

```python
for date in seluruh_tanggal: 
    print(date.text.strip())
```

```python
[seluruh_judul, seluruh_harga, seluruh_tanggal, seluruh_kota]
print(f'panjang seluruh judul: {len(seluruh_judul)}\n panjang seluruh harga: {len(seluruh_harga)}\n panjang seluruh tanggal: {len(seluruh_tanggal)}\n panjang seluruh kota: {len(seluruh_kota)}')
```

```python
paperwhite = []
for judul, harga, tanggal, kota in zip(seluruh_judul, seluruh_harga, seluruh_tanggal, seluruh_kota):
    item = {}
    print(judul.text.strip())
    item['judul']=judul.text.strip()
    print(harga.text.strip())
    item['harga'] = harga.text.strip()
    print(tanggal.text.strip())
    item['tanggal'] = tanggal.text.strip()
    print(kota.find('span').text.strip())
    item['kota'] = kota.find('span').text.strip()
    paperwhite.append(item)
```
